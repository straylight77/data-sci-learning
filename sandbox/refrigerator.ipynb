```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
url = "https://instruction.bus.wisc.edu/jfrees/jfreesbooks/regression%20modeling/bookwebdec2010/CSVData/Refrigerator.csv"
df = pd.read_csv(url)
df.info()
```

```python
pd.plotting.scatter_matrix(df, alpha=0.5)
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
ax = sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
df.plot.scatter(y="PRICE", x="FSIZE", alpha=0.5)
plt.show()
```

```python
df.plot.scatter(y="PRICE", x="FEATURES", alpha=0.5)
plt.show()
```

```python
sns.pairplot(df, kind="reg")
plt.show()
```

```python
#c = "Spectral_r"
#c = "coolwarm"
c = "vlag"
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap=c, vmin=-1, vmax=1, fmt=".2f", linewidth=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

# Load Data
iris = load_iris()

# Create a dataframe
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
#df.sample(4)
#print(iris.DESCR)
```

```python
df["species"] = df["target"].map( {0: "setosa", 1: "versicolor", 2: "virginica"} )
df.sample(5)
```

```python
df.info()
```

```python
df.groupby('target').size()
```

```python
cols_to_plot = [
    "sepal length (cm)", 
    "sepal width (cm)", 
    "petal length (cm)", 
    "petal width (cm)", 
    "species"
]
sns.pairplot(df[cols_to_plot], hue="species", markers=["o", "s", "D"], palette="Set2")
plt.show()
```

```python
sns.pairplot(df[cols_to_plot], kind="reg")
plt.show()
```

```python
df.boxplot(by="target")
plt.show()
```

```python
df.hist(column="petal length (cm)", grid=False, rwidth=0.9)
plt.show()
```

```python
sns.histplot(df, x="petal length (cm)", hue="species", kde=True)
```

```python
sns.kdeplot(df, x="petal length (cm)", hue="species", fill=True)
```
